Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for producing high-quality plots and figures in various formats such as line plots, scatter plots, bar plots, histograms, pie charts, and more. Matplotlib provides a MATLAB-like interface through its pyplot module, which makes it easy to get started with plotting in Python.


1.Line Plot: Line plots are used to display data points connected by straight line segments. They are commonly used to visualize trends over time or to show the relationship between two variables.

2.Scatter Plot: Scatter plots are used to visualize the relationship between two variables by displaying data points as individual markers on a two-dimensional plane.

3.Bar Plot: Bar plots are used to compare the values of different categories by displaying rectangular bars with heights proportional to the values they represent.


4.Histogram: Histograms are used to visualize the distribution of a continuous variable by dividing the data into bins and displaying the frequency of observations within each bin.

5.Pie Chart: Pie charts are used to represent the proportions of different categories as slices of a circular pie, where each slice's size is proportional to the value it represents.

6.scatter3d: To plot 3d graphs

 **2. What is a scatter plot? Use the following code to generate data for x and y.Using this generated data  plot a scatter plot.**

####import numpy as np
####np.random.seed(3)
####x = 3 + np.random.normal(0, 2, 50)
####y = 3 + np.random.normal(0, 2, len(x))
#####Note: Also add title, xlabel, and ylabel to the plot.**

ans. Scatter Plot: Scatter plots are used to visualize the relationship between two variables by displaying data points as individual markers on a two-dimensional plane.
python

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.normal(0,2,50)
y=np.random.normal(0,2,len(x))

plt.figure(figsize=(6,4))
plt.scatter(x=x,y=y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot x vs y')
```

** What is a box plot? Why is it used? Using the following data plot a box plot. **
###box1 = np.random.normal(100, 10, 200)
###box2 = np.random.normal(90, 20, 200)


A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It provides a compact way to visualize the central tendency, variability, and skewness of the data, as well as identifying potential outliers.

Box plots are particularly useful for comparing the distributions of multiple datasets or for identifying differences and similarities between groups of data. They are commonly used in exploratory data analysis and statistical analysis to gain insights into the underlying characteristics of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box=[box1,box2]
plt.figure(figsize=(5,10))
for i in range(len(box)):
  plt.subplot(2,1,1+i)
  plt.boxplot(box[i])
```
