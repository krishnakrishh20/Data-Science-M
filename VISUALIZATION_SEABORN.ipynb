**1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Seaborn is a powerful Python visualization library built on top of Matplotlib. It offers a variety of plots for exploring and visualizing data effectively. Here are five commonly used plots along with their uses:

####**Scatter Plot**:

 Scatter plots are useful for visualizing the relationship between two numerical variables. They help identify patterns, clusters, correlations, and outliers in data.
####**Histogram**:

 Histograms are used to represent the distribution of a single numerical variable. They help visualize the frequency or count of data within certain intervals or bins, showing the shape and spread of the data.
####**Bar Plot:**

Bar plots are effective for comparing categorical data. They display the distribution of a categorical variable by representing the frequency or mean of each category using bars. Bar plots are useful for visualizing counts, proportions, or aggregated values across categories.
####**Line Plot:**

Line plots are commonly used to visualize trends over time or any ordered variable. They connect data points with lines, making them suitable for displaying continuous data series such as time series data or sequences.
###**Box Plot**(Box-and-Whisker Plot):

Box plots are used for visualizing the distribution of numerical data through quartiles. They display the median, quartiles, and potential outliers of a dataset, making them useful for identifying the spread and central tendency of the data, as well as detecting any skewness or asymmetry.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
fmri_data = sns.load_dataset("fmri")

# Plot box plots
plt.figure(figsize=(6, 6))
sns.lineplot(data=fmri_data, x='timepoint', y='signal')
plt.title('Line plot of fmri')
plt.xlabel('timepoint')
plt.ylabel('signal')


plt.show()
```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

**6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Number of Passengers (Flights Dataset)')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
